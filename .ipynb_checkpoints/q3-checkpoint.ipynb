# Brute force mantığı
-Önce olası tüm item kombinasyonları belirlenmeli

bir perakende mağazasının satış verilerinde elma öğesi 1000 kez görüldüyse, elma öğesi için support değeri 1000 olur.

"eğer bir müşteri elma satın alırsa, o müşterinin portakal da satın alma olasılığı yüksektir" gibi bir ilişki kuralında, confident değeri, elma satın alan müşterilerin portakal da satın alma olasılığını gösterir. Confident değeri, 0 ile 1 arasında bir değer alır, ve 1'e ne kadar yakınsarsa ilişki kuralının doğruluğu o kadar yüksektir.

```python
min_sup=int(input("Enter minimum support: "))
min_conf=int(input("Enter minimum confidence: "))

veri = [] #minimum support değerinin üstündeki her bir sepet
          #kombinasyonunu ve supportcount değerlerini tutma
track = 0
print("Minimum support:", min_sup)
print("Minimum confident:", min_conf)
for i in range(1, len(unique_items)+1):
    
    if track%50==0:
        print(track)
        
    for olasi_sepet in combinations(unique_items, 1):
        
        support_count=0
        
        for sepet in data["item"]:
            
            if set(sepet).issubset(set(olasi_sepet)):
                
                support_count+=1
                
        if support_count>= min_sup:
            #print("zort")
            veri.append((sepet, support_count))
    track+=1
```

```python
# veri kümesi (mağaza satış verileri)
sales_data = data["item"].values

for i in range(1, len(unique_items)+1):
    for olasi_sepet in combinations(unique_items, 1):
        item_sets.append([olasi_sepet])

# ilişki kurallarını bulma
rules = []
for item_set in item_sets:
    for sale in sales_data:
        if all(item in sale for item in item_set):
            for item in sale:
                if item not in item_set:
                    rules.append((item_set, [item]))

# support ve confident değerlerini hesaplama
for rule in rules:
    item_set, conclusion = rule
    support = 0
    confident = 0
    for sale in sales_data:
        if all(item in sale for item in item_set):
            support += 1
        if all(item in sale for item in item_set + conclusion):
            confident += 1
    support = confident / len(sales_data)
    confident = confident / support
    print("Kural: ", item_set, "=>", conclusion)
    print("Support: ", support)
    print("Confidence: ", confident)
```

```python
# veri kümesi (mağaza satış verileri)
sales_data = [
    ['elma', 'portakal', 'muz'],
    ['elma', 'muz'],
    ['elma', 'portakal', 'muz'],
    ['elma', 'portakal', 'kivi'],
    ['elma', 'kivi'],
    ['portakal', 'muz'],
]

# tüm öğe kümelerini bulma
item_sets = []
for sale in sales_data:
    for item in sale:
        item_sets.append([item])

# ilişki kurallarını bulma
rules = []
for item_set in item_sets:
    for sale in sales_data:
        if all(item in sale for item in item_set):
            for item in sale:
                if item not in item_set:
                    rules.append((item_set, [item]))

# support ve confident değerlerini hesaplama
for rule in rules:
    item_set, conclusion = rule
    support = 0
    confident = 0
    for sale in sales_data:
        if all(item in sale for item in item_set):
            support += 1
        if all(item in sale for item in item_set + conclusion):
            confident += 1
    support = confident / len(sales_data)
    confident = confident / support
    print("Kural: ", item_set, "=>", conclusion)
    print("Support: ", support)
    print("Confidence: ", confident)
```
